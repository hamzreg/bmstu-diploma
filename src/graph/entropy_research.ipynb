#### Оптимизация метода скользящего окна

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

DATA_PATH = 'https://raw.githubusercontent.com/hamzreg/bmstu-diploma/main/src/scripts/stat/vm2/stat.csv'

df = pd.read_csv(DATA_PATH)


# матрица корреляции
data = df[['KH_ENTROPY', 'COMPRESSION_RATIO', 'SPACE_SAVINGS', 'COMPRESSION_FACTOR', 'COMPRESSION_GAIN']]
matrix = data.corr()
sns.heatmap(matrix, annot=True, vmax=1, vmin=-1, center=0, cmap='vlag')


# логарифмирование значений по X и Y
df['LOG_KH_ENTROPY'] = np.log(df['KH_ENTROPY'])
df['LOG_COMPRESSION_RATIO'] = np.log(df['COMPRESSION_RATIO'])
df['LOG_SPACE_SAVINGS'] = np.log(df['SPACE_SAVINGS'])
df['LOG_COMPRESSION_FACTOR'] = np.log(df['COMPRESSION_FACTOR'])
df['LOG_COMPRESSION_GAIN'] = np.log(df['COMPRESSION_GAIN'])


# для построения trendline
z1 = np.polyfit(df['LOG_KH_ENTROPY'], df['LOG_COMPRESSION_RATIO'], 1)
p1 = np.poly1d(z1)

z2 = np.polyfit(df['LOG_KH_ENTROPY'], df['LOG_SPACE_SAVINGS'], 1)
p2 = np.poly1d(z2)

z3 = np.polyfit(df['LOG_KH_ENTROPY'], df['LOG_COMPRESSION_FACTOR'], 1)
p3 = np.poly1d(z3)

z4 = np.polyfit(df['LOG_KH_ENTROPY'], df['LOG_COMPRESSION_GAIN'], 1)
p4 = np.poly1d(z4)


# построение диаграмм рассеяния
fig1, axs1 = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
df.plot(ax=axs1[0], x='KH_ENTROPY', y='COMPRESSION_RATIO', kind='scatter', s=5)
df.plot(ax=axs1[1], x='LOG_KH_ENTROPY', y='LOG_COMPRESSION_RATIO', kind='scatter', s=5)
axs1[2].scatter(df['LOG_KH_ENTROPY'], df['LOG_COMPRESSION_RATIO'], s=5)
axs1[2].plot(df['LOG_KH_ENTROPY'], p1(df['LOG_KH_ENTROPY']), "r--", linewidth=1)

fig2, axs2 = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
df.plot(ax=axs2[0], x='KH_ENTROPY', y='SPACE_SAVINGS', kind='scatter', s=5, color='g')
df.plot(ax=axs2[1], x='LOG_KH_ENTROPY', y='LOG_SPACE_SAVINGS', kind='scatter', s=5, color='g')
axs2[2].scatter(df['LOG_KH_ENTROPY'], df['LOG_SPACE_SAVINGS'], s=5, c='g')
axs2[2].plot(df['LOG_KH_ENTROPY'], p2(df['LOG_KH_ENTROPY']), "r--", linewidth=1)

fig1, axs3 = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
df.plot(ax=axs3[0], x='KH_ENTROPY', y='COMPRESSION_FACTOR', kind='scatter', s=5, color='m')
df.plot(ax=axs3[1], x='LOG_KH_ENTROPY', y='LOG_COMPRESSION_FACTOR', kind='scatter', s=5, color='m')
axs3[2].scatter(df['LOG_KH_ENTROPY'], df['LOG_COMPRESSION_FACTOR'], s=5, c='m')
axs3[2].plot(df['LOG_KH_ENTROPY'], p3(df['LOG_KH_ENTROPY']), "r--", linewidth=1)

fig1, axs4 = plt.subplots(nrows=1, ncols=3, figsize=(20, 5))
df.plot(ax=axs4[0], x='KH_ENTROPY', y='COMPRESSION_GAIN', kind='scatter', s=5, color='y')
df.plot(ax=axs4[1], x='LOG_KH_ENTROPY', y='LOG_COMPRESSION_GAIN', kind='scatter', s=5, color='y')
axs4[2].scatter(df['LOG_KH_ENTROPY'], df['LOG_COMPRESSION_GAIN'], s=5, c='y')
axs4[2].plot(df['LOG_KH_ENTROPY'], p4(df['LOG_KH_ENTROPY']), "r--", linewidth=1)
```
